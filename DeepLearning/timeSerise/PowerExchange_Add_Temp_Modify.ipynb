```python
import pandas as pd
```

```python
data_dir = '/Volumes/Data/Row_Data'
load_file = 'kpx_power_demand_addTemp_2023.csv'
```

```python
url = f'{data_dir}/{load_file}'
df_data = pd.read_csv(url,  parse_dates=['date_time'], index_col = 'date_time' )
df_data
```

```python
df_data = pd.get_dummies(df_data, columns = ['month', 'weekday', 'hr', 'temp_boundary'])
df_data = df_data.drop( columns=['base_time', 'temp', 'day'] )
df_data
```
