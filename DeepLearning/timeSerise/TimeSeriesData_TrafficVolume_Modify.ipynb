```python
def weather_main_modify(x):
    rain = ['Rain', 'Thunderstorm', 'Squall', 'Drizzle']
    fog = ['Fog', 'Mist', 'Haze', 'Smoke']
    snow = ['Snow']
    clear = ['Clear', 'Clouds']

    if x in rain:
        return 'Rain'
    elif x in fog:
        return 'Fog'
    elif x in snow:
        return 'Snow'
    else: 
        return 'Clear'
```

```python
df_dummy['weather_main'] = df_dummy.weather_main.map(weather_main_modify)
df_dummy.head()
```

```python
df_traffic_volume_mean = df_dummy.groupby(['holiday', 'weekday','hour']).aggregate( {'traffic_volume':'mean'})
df_traffic_volume_mean
```

```python
df_traffic_modify = df_dummy.loc[ df_dummy.traffic_volume.isna()].copy()
df_traffic_modify                                                               
```

```python
for index in range(len(df_traffic_modify)):
    row_data = df_traffic_modify.iloc[index]
    traffic_volumn = df_traffic_volume_mean.xs( key=(row_data.holiday, row_data.weekday, row_data.hour) ).values[0]
    df_traffic_modify.loc[df_traffic_modify.date_time == row_data.date_time, 'traffic_volume'] = round(traffic_volumn)

df_traffic_modify.head()
```

```python
df_dummy.update(df_traffic_modify, overwrite=False)
df_dummy.head()
```

```python
def hour_modify(x):
    middle_volume = [5,19, 20, 21, 22]
    small_volume = [23,0,1,2,3,4]
    if x in small_volume:
        return 'SmallVolume'
    elif x in middle_volume:
        return 'MiddleVolume'
    else:
        return 'BigVolume'
```

```python
df_dummy['hour'] = df_dummy['hour'].map(hour_modify)
df_dummy.head()
```
