```python
import pandas as pd
```

```python
data_dir = '/Volumes/Data/Row_Data'
file_name = 'kpx_power_demand_2013_2023.csv'
file_name_modify = 'kpx_powerDemand_2013_2023_v1.csv'
url = f'{data_dir}/{file_name}'
df = pd.read_csv(url, encoding='euc-kr')
df
```

```python
# 누락 데이터 검증 
date_range = pd.date_range(df.날짜.min(), df.날짜.max(), freq='D')
date_range
```

```python
# 중복 데이터 검증 
df_test = df[~df.날짜.duplicated(keep='last')]
df_test.shape
```

```python
columns = list(range(24))
columns.insert(0, 'baseDatetime')
df.columns = columns
df
```

```python
df.melt(id_vars=['baseDatetime'])
```

```python
df_data = df.melt(id_vars=['baseDatetime']).sort_values(by=['baseDatetime', 'variable'])
df_data.columns = ['baseDatetime', 'hr', 'demand']
df_data
```

```python
df_data.index = df_data[['baseDatetime', 'hr']].apply( lambda x:  pd.to_datetime(  '%s %02d:00:00' % (x[0], x[1]),  format='%Y-%m-%d %H:%M:%S' )     , axis=1) 
df_data
```

```python
df_data['baseDatetime'] = df_data.index
df_data['weekday'] = df_data['baseDatetime'].dt.weekday
df_data['month'] = df_data['baseDatetime'].dt.month
df_data = df_data.drop( columns= [ 'baseDatetime'])
df_data = df_data[[ 'month', 'weekday', 'hr', 'demand']]
df_data
```

```python
df_data.info()
```

```python
def weekday_modify(x):
    large_volume = [0,1,2,3,4]
    middle_volume = [5]
    small_volume = [6]
    
    if x in small_volume:
        return 'SmallVolume'
    elif x in middle_volume:
        return 'MiddleVolume'
    else:
        return 'LargeVolume'


df_data['weekday'] = df_data['weekday'].map(weekday_modify)
df_data
```

```python
df_dummy = pd.get_dummies(df_data, columns = ['month', 'weekday', 'hr'])
df_dummy.head()
```
