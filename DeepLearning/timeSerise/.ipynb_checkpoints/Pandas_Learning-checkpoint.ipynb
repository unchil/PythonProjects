```python
import pandas as pd
import numpy as np
```

```python
def make_timeseries(start='2013-01-01', end='2014-01-01', freq='1D', seed=None):
    index = pd.date_range(start=start, end=end, freq=freq, name='timestamp')
    row_cnt = len(index)
    state = np.random.RandomState(seed)
    data = {
        'name':state.choice(["Alice", "Bob", "Charlie"], size=row_cnt),
        'id':state.poisson(1000, size=row_cnt),
        'x':state.rand(row_cnt) * 2 -1,
        'y':state.rand(row_cnt) * 2 -1   
    }
    df = pd.DataFrame(data, index=index, columns=sorted(data))

    return df.iloc[:-1]
```

```python
timeseries = [
    make_timeseries(freq="1min", seed=i).rename(columns=lambda x: f"{x}_{i}")
    for i in range(10)
]
```

```python
ts = make_timeseries(freq="30s", seed=0)
ts.to_parquet("timeseries.parquet")
ts = pd.read_parquet("timeseries.parquet")
ts
```

```python
ts.info()
```

```python
ts.memory_usage(deep=True)
```

```python
ts2 = ts.copy()
ts2.name = ts2.name.astype('category')
ts2.info()
```

```python
ts2.memory_usage(deep=True)
```

```python
ts2.id = pd.to_numeric(ts2.id, downcast='unsigned')
ts2[['x', 'y']] = ts2[['x', 'y']].apply(pd.to_numeric, downcast='float')
ts2.info()
```

```python
ts2.memory_usage(deep=True)
```

```python
reduction =( ts.memory_usage(deep=True).sum() - ts2.memory_usage(deep=True).sum() ) /1024/1024
print(f"Raw Mem:{ts.memory_usage(deep=True).sum()/1024/1024:0.2f}MB, DownCast Mem:{ts2.memory_usage(deep=True).sum()/1024/1024:0.2f}MB, Reduction:{reduction:0.2f}MB")
```
