```python
df_data = df_data.drop( columns = ['연번', '역번호'] )
df_data
```

```python
df_data = df_data.melt(
    id_vars=[ '요일구분', '호선', '출발역', '상하구분']
)
df_data
```

```python
start = '2024-07-24 05:30'
end = '2024-07-25 00:30 '
index = pd.date_range(start, end, freq='30min')
column_times = ['5시30분', '6시00분', '6시30분',
       '7시00분', '7시30분', '8시00분', '8시30분', '9시00분', '9시30분', '10시00분',
       '10시30분', '11시00분', '11시30분', '12시00분', '12시30분', '13시00분', '13시30분',
       '14시00분', '14시30분', '15시00분', '15시30분', '16시00분', '16시30분', '17시00분',
       '17시30분', '18시00분', '18시30분', '19시00분', '19시30분', '20시00분', '20시30분',
       '21시00분', '21시30분', '22시00분', '22시30분', '23시00분', '23시30분', '00시00분',
       '00시30분']

def transCollectTime(x):
    for i, value in enumerate(column_times):
        if x == value:
            return index[i]
```

```python
df_data['time'] = df_data.variable.map(transCollectTime)
#data = data.sort_values(by=['출발역', '호선', '상하구분', '요일구분','time'], ascending=True )
df_data = df_data.drop( columns = ['variable'] )
df_data.index = df_data.time
df_data = df_data.drop( columns = ['time'] )
df_data
```

```python
df_data = df_data.rename(columns={'value':'혼잡도'})
df_data
```
