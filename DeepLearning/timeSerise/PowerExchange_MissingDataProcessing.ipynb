```python
import pandas as pd
```

```python
data_dir = '/Volumes/Data/Row_Data'
load_file = 'kpx_powerDemand_2013_2023_v2.csv'
url = f'{data_dir}/{load_file}'

df_data = pd.read_csv(url, parse_dates=['date_time'], index_col='date_time')
df_data
```

```python
df_data.temp.ffill(inplace=True)
df_data.temp.info()
```

```python
def temp_boundary(x):
    if x <=4:
        return 1      
    elif x > 4 and x <=8:
        return 2
    elif x > 8 and x <=24:
        return 3 
    elif x > 24 and x <= 28:
        return 4    
    elif x > 28 and x <= 32:
        return 5    
    else:
        return 6


df_data['temp'] = df_data['temp'].map(temp_boundary)
df_data
```

```python
df_data.info()
```

```python

data_dir = '/Volumes/Data/Row_Data'
save_file = 'kpx_powerDemand_2013_2023_v3.csv'
url = f'{data_dir}/{save_file}'

df_data.to_csv(url)  
```
