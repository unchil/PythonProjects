```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_weather_data = pd.read_csv('./data/london_weather.csv',  parse_dates=['date'])
df_weather_data.head()
```

```python
df_weather_data = df_weather_data.loc[df_weather_data.date.between('2012-01-01', '2013-12-31')]
df_weather_data.shape
```

```python
df_weather_data
```

```python
df_weather_data['snow'] = df_weather_data['snow_depth'].map(lambda x: False if x == 0 else True)
df_weather_data
```

```python
df_weather_data['rain'] = df_weather_data['precipitation'].map(lambda x: False if x == 0 else True)
df_weather_data
```

```python
def weather_maxtemp_modify(x):
    if x < 11:
        return '1'
    elif x >= 11 and x < 15:
        return '2'
    elif x >= 15 and x < 20:
        return '3' 
    else: 
        return '4'
                                                                                    
def weather_mintemp_modify(x):
    if x < 4:
        return '1'
    elif x >= 4 and x < 8:
        return '2'
    elif x >= 8 and x < 12:
        return '3'    
    else: 
        return '4'

def weather_meantemp_modify(x):
    if x < 7:
        return '1'
    elif x >= 7 and x < 11:
        return '2'
    elif x >= 11 and x < 16:
        return '3'    
    else: 
        return '4'
```

```python
df_weather_data['temp_max'] = df_weather_data.max_temp.map(weather_maxtemp_modify)
df_weather_data
```

```python
df_weather_data['temp_min'] = df_weather_data.min_temp.map(weather_mintemp_modify)
df_weather_data
```

```python
df_weather_data['temp_mean'] = df_weather_data.mean_temp.map(weather_mintemp_modify)
df_weather_data
```

```python
df_weather_data = df_weather_data.drop(columns=['sunshine','global_radiation','pressure','cloud_cover', 'precipitation', 'snow_depth', 'mean_temp', 'min_temp','max_temp'])
df_weather_data.shape
```

```python
df_weather_data
```

```python
df_weather_data['month'] = df_weather_data.date.dt.month
df_weather_data
```

```python
df_weather_data.index = df_weather_data.date
```

```python
df_weather_data
```

```python
df_weather_data = df_weather_data.drop(columns=['date'])
df_weather_data
```

```python
df_weather_data = pd.get_dummies(df_weather_data, columns = ['snow', 'rain', 'temp_max', 'temp_min', 'temp_mean', 'month'])
df_weather_data.to_csv('./data/london_weather_modify_v2.csv')
```
