```python
data_path = '/Users/unchil/PythonProjects/data/AirQualCollector/AirQualCollector_2019_12.csv'

df_data = pd.read_csv(data_path,   parse_dates=['TimeStamp'])
df_data.head()
```

```python
df_data = df_data.loc[ (df_data.Station_ID == 1002) ]
```

```python
df_data = df_data.drop(columns= ['Station_ID', 'AirQualCollectorID', 'AirQualControllerID', 'CollectorModel', 'IF_KEY'])
df_data.shape
```

```python
pd.date_range(df_data.TimeStamp.min(), df_data.TimeStamp.max(), freq='s').shape
```

```python
def makeData(df_data):

    dummy_list = []
    
    df_data.TimeStamp = df_data.TimeStamp.map(lambda x: x.replace( microsecond=0)) 
    time_serise_index = pd.date_range(df_data.TimeStamp.min(), df_data.TimeStamp.max(), freq='s')
 
    for location in df_data.Location.unique():
        df_data_location = df_data.loc[ (df_data.Location == location) ]
        df_data_location = df_data_location.loc[~df_data_location.TimeStamp.duplicated(keep='last')]
        df_data_location.index = df_data_location.TimeStamp
       # df_data_location = df_data_location.sort_index(ascending=True)

        df_data_location_dummy = pd.DataFrame(data=None, columns=df_data_location.columns, index = time_serise_index)
        df_data_location_dummy.TimeStamp = df_data_location_dummy.index
        df_data_location_dummy.update(df_data_location, overwrite=False)
        df_data_location_dummy = df_data_location_dummy.ffill()
        dummy_list.append(df_data_location_dummy)
        del(df_data_location_dummy)
        
    return_val = pd.concat(dummy_list)
    del(dummy_list)
    
    return return_val
```

```python
df_data_new = makeData(df_data) 
df_data_new.shape
```
