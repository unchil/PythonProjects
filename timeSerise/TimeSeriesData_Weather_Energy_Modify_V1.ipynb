```python
import pandas as pd
```

```python
df_weather_data = pd.read_csv('./data/london_weather_modify_v2.csv',  parse_dates=['date'], index_col= 'date')
df_weather_data
```

```python
df_energy_train = pd.read_csv('./data/london_energy_train.csv',  parse_dates=['date'], index_col= 'date')
df_energy_train
```

```python
df_energy_valid = pd.read_csv('./data/london_energy_valid.csv',  parse_dates=['date'], index_col='date')
df_energy_valid
```

```python
df_energy_test = pd.read_csv('./data/london_energy_test.csv',  parse_dates=['date'], index_col='date')
df_energy_test
```

```python
df_weather_energy_train = df_energy_train.join(df_weather_data, how='left')
df_weather_energy_train
```

```python
df_weather_energy_valid = df_energy_valid.join(df_weather_data, how='left')
df_weather_energy_valid
```

```python
df_weather_energy_test = df_energy_test.join(df_weather_data, how='left')
df_weather_energy_test
```

```python
df_weather_energy_train.to_csv('./data/london_weather_energy_train.csv')
df_weather_energy_valid.to_csv('./data/london_weather_energy_valid.csv')
df_weather_energy_test.to_csv('./data/london_weather_energy_test.csv')
```
