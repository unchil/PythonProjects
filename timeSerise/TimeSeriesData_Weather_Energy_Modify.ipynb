```python
import pandas as pd
```

```python
df_weather_energy = pd.read_csv('./data/london_weather_energy.csv',  parse_dates=['date'])
df_weather_energy
```

```python
df_weather_energy['snow'] = df_weather_energy['snow_depth'].map(lambda x: False if x == 0 else True)
df_weather_energy
```

```python
df_weather_energy['rain'] = df_weather_energy['precipitation'].map(lambda x: False if x == 0 else True)
df_weather_energy
```

```python
def weather_maxtemp_modify(x):
    if x < 11:
        return '1'
    elif x >= 11 and x < 15:
        return '2'
    elif x >= 15 and x < 20:
        return '3' 
    else: 
        return '4'
                                                                                    
def weather_mintemp_modify(x):
    if x < 4:
        return '1'
    elif x >= 4 and x < 8:
        return '2'
    elif x >= 8 and x < 12:
        return '3'    
    else: 
        return '4'

def weather_meantemp_modify(x):
    if x < 7:
        return '1'
    elif x >= 7 and x < 11:
        return '2'
    elif x >= 11 and x < 16:
        return '3'    
    else: 
        return '4'
```

```python
df_weather_energy['temp_max'] = df_weather_energy.max_temp.map(weather_maxtemp_modify)
df_weather_energy
```

```python
df_weather_energy['temp_min'] = df_weather_energy.min_temp.map(weather_mintemp_modify)
df_weather_energy
```

```python
df_weather_energy['temp_mean'] = df_weather_energy.mean_temp.map(weather_mintemp_modify)
df_weather_energy
```

```python
df_weather_energy = df_weather_energy.drop(columns=['cloud_cover', 'precipitation', 'snow_depth', 'mean_temp', 'min_temp','max_temp', 'min'])
df_weather_energy.shape
```

```python
df_weather_energy
```

```python
df_weather_energy['month'] = df_weather_energy.date.dt.month
df_weather_energy
```

```python
df_weather_energy.index = df_weather_energy.date
```

```python
df_weather_energy
```

```python
df_weather_energy = df_weather_energy.drop(columns=['date'])
df_weather_energy
```

```python
df_weather_energy = pd.get_dummies(df_weather_energy, columns = ['snow', 'rain', 'temp_max', 'temp_min', 'temp_mean', 'month'])
```
